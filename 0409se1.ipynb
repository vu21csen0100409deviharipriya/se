```python
import matplotlib.pyplot as plt
import numpy as np

# Hard-coded coefficients for the quadratic equation (ax^2 + bx + c)
a = 0.1
b = -2
c = 10

# Generate x values (days)
days = np.arange(1, 31, 1)

# Calculate y values using the quadratic equation
temperature = a * days**2 + b * days + c

# Plot the graph
plt.plot(days, temperature, label='Weather Model')

# Add labels and title
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Quadratic Weather Model')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def generate_weather_plot(days, temp_coefficient, rain_coefficient, initial_temp, initial_rain, label):
    days_array = np.arange(1, int(days) + 1)
    temperature = initial_temp - temp_coefficient * (days_array - 1) ** 2
    rainfall = initial_rain + rain_coefficient * (days_array - 1) ** 2

    plt.plot(days_array, temperature, label=f'Temperature {label}')
    plt.plot(days_array, rainfall, label=f'Rainfall {label}')

def main():
    # Predefined sets of values as a list of lists
    parameter_sets = [
        [10, 2.5, 0.5, 25.0, 5.0],
        [15, 3.0, 0.6, 22.0, 7.0],
        # Add more sets as needed
    ]

    # Process each set of parameters
    for i, parameters in enumerate(parameter_sets):
        generate_weather_plot(*parameters, label=f'Set {i + 1}')

    # Adding labels and title
    plt.xlabel('Days')
    plt.ylabel('Value')
    plt.title('Weather Modeling with Quadratic Solution')
    plt.legend()

    # Displaying the plot
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Read coefficients from a text file
with open('coefficients.txt', 'r') as file:
    a, b, c = map(int, file.readline().strip().split())

# Generate x values (days)
days = np.arange(1, 31, 1)

# Calculate y values using the quadratic equation
temperature = a * days**2 + b * days + c

# Plot the graph
plt.plot(days, temperature, label='Weather Model')

# Add labels and title
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Quadratic Weather Model')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
